```python
import anndata
import pandas as pd
import scgenome.plotting as pl
import os
```

```python
# file_path = "/Users/zemp/Downloads/data/signals_SPECTRUM-OV-014.h5"
# file_path = "/Users/zemp/Downloads/data/signals_SPECTRUM-OV-006.h5"
file_path = "/Users/zemp/Downloads/data/signals_SPECTRUM-OV-107.h5"
ad = anndata.read_h5ad(file_path)
g = pl.plot_cell_cn_matrix_fig(ad, layer_name='copy', raw=True, cell_order_fields=['cluster_id'], annotation_fields=['cluster_id'])
```

```python
x_df = ad.transpose().to_df('copy')
pos_df = ad.var[['chr', 'start', 'end']]
dat_df = pd.concat([pos_df, x_df], axis=1)
dat_df.iloc[:,:20].head()
```

```python
filename, extension = os.path.splitext(file_path)
dat_df.to_csv(filename + "_corr.csv.gz", index=False, compression='gzip')
```

```python
# save raw reads
ad.layers['raw'] = ad.X
raw_df = pd.concat([pos_df, ad.transpose().to_df('raw')], axis=1)
raw_df.iloc[:, :20].head()
```

```python
raw_df.to_csv(filename + ".csv.gz", index=False, compression='gzip')
```
